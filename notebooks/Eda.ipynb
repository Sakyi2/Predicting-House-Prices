```python
#visualie the numberical features of the distribution
num_f =df.select_dtypes(include=['int64' ,'float64']).columns


#ploting one column at a time 
for col in num_f:
  plt.figure(figsize=(3,4)),
  sns.histplot(df[col],kde=True)#kda adds a smooth line to the hist
  plt.title('numberical features distribution {col}')
plt.show()
```
